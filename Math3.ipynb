```python
Задача 3:
Известно, что в 2024 году средний балл за ЕГЭ по профильной математике составил 62.6 со стандартным отклонением 6.4. 
В некотором классе 20 учеников написали экзамен таким образом, что их средний результат оказался равным 66.1. 
На уровне статистической значимости α = 0.05 проверьте гипотезу H1 = ” Этот класс написал значительно лучше, чем в среднем по стране ”.

Сформулируем гипотезы:
    - Нулевая гипотеза H0: Средний результат класса не превышает средний по стране mu = 62.6.
    - Альтернативная гипотеза H1: Средний результат класса существенно лучше mu > 62.6.
```

```python
# Данные задачи
mu_country = 62.6  # Средний балл по стране
sigma_country = 6.4  # Стандартное отклонение по стране
mean_class = 66.1  # Средний балл класса
n_class = 20  # Число учеников в классе
alpha = 0.05  # Уровень значимости

# Шаг 1: Рассчитываем стандартную ошибку среднего
standard_error = sigma_country / (n_class ** 0.5)
print(f"Стандартная ошибка: {standard_error:.2f}")

# Шаг 2: Рассчитать t-статистику
t_statistic = (mean_class - mu_country) / standard_error
print(f"t-статистика: {t_statistic:.4f}")

# Шаг 3: Критическое значение t
# При уровне значимости 0.05 и 19 степенях свободы, критическое значение t ≈ 1.729
t_critical = 1.729

# Шаг 4: Сравнить t-статистику с критическим значением
if t_statistic > t_critical:
    print("Отклоняем нулевую гипотезу: класс написал значительно лучше, чем в среднем по стране.")
else:
    print("Не удалось отклонить нулевую гипотезу: нет статистически значимых доказательств, что класс написал лучше.")
```

```python
Задача 4
Поставщик кофе утверждает, что средний вес пачки кофе составляет ровно 100 г. 
Во время контрольной закупки из партии извлекается выборка из десяти пачек и взвешивается. 
Массы пачек (в граммах) равны: 97, 98, 104, 97, 101, 99, 99, 101, 98 и 100. 
Можно ли на уровне 99% доверять поставщику относительно средней массы пачки? 
Считайте, что вес пачки распределен нормально.

Сформулируем гипотезы:
    - Нулевая гипотеза H0: редний вес пачки кофе составляет ровно 100 г.
    - Альтернативная гипотеза H1: Средний вес пачки кофе составляет не равен 100 г.
```

```python
# Данные выборки
weights = [97, 98, 104, 97, 101, 99, 99, 101, 98, 100]

# Параметры гипотезы
mu_hypothesized = 100  # Средняя масса, заявленная поставщиком
alpha = 0.01  # Уровень значимости

# Шаг 1: Вычисление выборочного среднего
mean_sample = sum(weights) / len(weights)
print(f"Среднее значение выборки: {mean_sample:.2f}")

# Шаг 2: Вычисление несмещенного стандартного отклонения
# 1. Вычисляем сумму квадратов отклонений от среднего
sum_squared_deviations = sum((x - mean_sample) ** 2 for x in weights)

# 2. Разделяем на (n-1) для несмещенного оценивания
std_sample = (sum_squared_deviations / (len(weights) - 1)) ** 0.5
print(f"Несмещенное стандартное отклонение: {std_sample:.2f}")

# Шаг 3: Рассчитываем t-статистику 
n = len(weights)  # Размер выборки
standard_error = std_sample / (n ** 0.5)  # стандартная ошибка среднего
t_statistic = (mean_sample - mu_hypothesized) / standard_error
print(f"t-статистика: {t_statistic:.4f}")

# Шаг 4: Сравниваем с критическим значением t
# Для 99% доверия и 9 степеней свободы (n-1) критическое значение из таблицы составляет примерно ±3.250
t_critical = 3.250

if abs(t_statistic) > t_critical:
    print("Отклоняем нулевую гипотезу: средний вес пачки статистически отличается от 100 г.")
else:
    print("Не удалось отклонить нулевую гипотезу: нет статистически значимых доказательств, что средний вес пачки отличается от 100 г.")
```
